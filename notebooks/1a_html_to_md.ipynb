```python
from markdownify import markdownify as md
import os
import re
from bs4 import BeautifulSoup
```

```python
for i in os.listdir("../data/0_raw_html"):
    print("Processing:", "http://www.cibc.com" + i.replace("_", "/"))
    try:
        with open(f"../data/0_raw_html/{i}", "r", encoding="utf-8") as f:
            html = f.read()
    except UnicodeDecodeError as e:
        print("UnicodeDecodeError:", "http://www.cibc.com" + i.replace("_", "/"))
        continue

    soup = BeautifulSoup(html, "html.parser")
    content = soup.find("main")
    if content is not None:
        content = content.prettify()
        content = md(content, strip=['a', "href", "img", "btn"])
        #replaces 3+ newlines with 2 newlines
        content = re.sub(r"\n{3,}", "\n\n", content)
        with open(f"../data/1_html_to_md/{i}", "w") as f:
            f.write(md(content))
```
